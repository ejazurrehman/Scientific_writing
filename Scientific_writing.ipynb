# What is Scientific Writing?

Scientific writing is a formal style of writing used to communicate research, experiments, and technical ideas in a clear, structured, and objective way. It is the standard mode of communication in disciplines like biology, physics, medicine, engineering, and social sciences.

Core idea:

At its core, scientific writing is about presenting evidence-based information so that others can understand, evaluate, and replicate it.

## Key Characterstics:

1. Clarity and precision
The language is exact and unambiguous. Vague terms are avoided. For example, instead of saying “a lot of increase,” you specify “a 23% increase.”

2. Objectivity
The focus is on data and evidence, not personal opinions. Writers avoid emotional or subjective language.

3. Structured format
Most scientific writing follows a standard structure, often called IMRAD:

- Introduction: What is the problem?

- Methods: How was the study conducted?

- Results: What was found?

- Discussion: What do the findings mean?

4. Evidence-based
All claims must be supported by data, experiments, or credible references.

5. Formal tone
It avoids casual language, slang, or storytelling. Sentences are direct and professional.

## Common types of scientific writing:

- Research papers (journal articles)

- Literature reviews

- Case studies

- Technical reports

- Theses and dissertations

## Why it matters:

Scientific writing is critical because it:

- Enables knowledge sharing across the global research community

- Allows others to verify and replicate findings

- Supports evidence-based decision-making in policy, medicine, and industry

## Standard structure of scientific writing:

### 1. Title

A concise statement of what the study is about.
It should be specific, not vague.

Example:
“Impact of Supply Chain Disruptions on Firm Performance in Emerging Markets”

### 2. Abstract

A brief summary of the entire study (usually 150–250 words).

It typically includes:

- Purpose of the study

- Method used

- Key results

- Main conclusion

**Think of it as a “mini version” of the whole paper.**

### 3. Introduction

This section sets the context.

It answers:

- What is the problem?

- Why is it important?

- What is already known?

- What gap does your study address?

- What are your objectives or hypotheses?

**In our field, this is where we might frame a real-world issue like supply chain disruptions or financial volatility.**

### 4. Literature Review (sometimes separate, sometimes inside Introduction)

This reviews existing research.

Purpose:

- Show what others have done

- Identify gaps or inconsistencies

- Justify your study

**It should not just summarize papers, but critically analyze them.**

### 5. Methodology (Methods)

This explains how the study was conducted.

Includes:

- Research design (qualitative, quantitative, mixed)

- Data sources (e.g., firm-level data, surveys)

- Variables and measurements

- Analytical techniques (regression, models, etc.)

**The key requirement: someone else should be able to replicate our study.**

### 6. Results

This section presents findings only, without interpretation.

- Tables, graphs, statistical outputs

- Clear reporting of outcomes

- Avoid explaining “why” here, just state “what”.

### 7. Discussion

Now we interpret the results.

- What do the findings mean?

- Do they support our hypotheses?

- How do we compare with past studies?

- What are the implications (theoretical or practical)?

**This is where our analytical thinking shows.**

### 8. Conclusion

A concise wrap-up.

Includes:

- Key findings

- Contributions

- Limitations

- Suggestions for future research

### 9. References

A list of all sources cited, formatted in a specific style (APA, Harvard, etc.).

### 10. Appendices (optional)

- Extra material:

- Questionnaires

- Detailed tables

- Technical derivations

## Quick mental model:

We can think of the structure like this:

- Introduction → What is the problem?

- Literature → What do we know?

- Method → What did we do?

- Results → What did we find?

- Discussion → What does it mean?

# Standard structure of a PhD thesis

A PhD thesis follows a more extensive and rigorous structure than a standard research paper. While formats vary slightly by university and discipline, most theses follow a multi-chapter framework built around the same scientific logic but with deeper justification and analysis.

### 1. Preliminary pages (front matter)

These are formal but important.

- Title page

- Declaration (originality statement)

- Certificate (supervisor approval)

- Acknowledgements

- Abstract (often longer than a paper abstract)

- Table of contents

- List of figures and tables

### 2. Chapter 1: Introduction

This is not just background. It sets up the entire thesis.

Typically includes:

- Research problem

- Research objectives

- Research questions or hypotheses

- Significance of the study

- Scope and limitations

- Structure of the thesis

**Think of this as our “research blueprint.”**

### 3. Chapter 2: Literature Review

Much deeper than a paper.

- Theoretical foundations

- Review of key studies

- Critical analysis (not just summary)

- Identification of research gaps

- Development of conceptual framework

**In finance or supply chain, this often includes models, theories, and competing perspectives.**

### 4. Chapter 3: Methodology

Highly detailed and justified.

Includes:

- Research philosophy (positivist, interpretivist, etc.)

- Research design

- Data sources and sampling

- Variable definitions and measurements

- Model specification (e.g., regression equations)

- Tools and software used

- Validity and reliability

**Examiners expect us to defend why each method was chosen.**

### 5. Chapter 4: Data Analysis / Results

Presentation of findings.

- Descriptive statistics

- Model outputs (regressions, SEM, etc.)

- Tables, figures, charts

**Clear, systematic, and aligned with research questions.**

### 6. Chapter 5: Discussion

This is where many theses succeed or fail.

- Interpretation of results

- Link back to theory

- Comparison with prior studies

- Explanation of unexpected findings

**This is not repetition. It is analytical synthesis.**

### 7. Chapter 6: Conclusion and Recommendations

Wraps everything together.

- Summary of key findings

- Contributions (theoretical and practical)

- Policy or managerial implications

- Limitations

- Future research directions

### 8. References

Comprehensive and strictly formatted.

### 9. Appendices

Supporting material such as:

- Survey instruments

- Additional tables

- Code or technical details

## Two common thesis formats

### 1. Traditional (Monograph)

- One continuous document

- Common in social sciences and management

### 2. Thesis by Publication

- Collection of published or publishable papers

- Each paper acts as a chapter

- Includes an integrating introduction and conclusion

**In our domain, many universities now encourage this format.**

# How a PhD thesis differs from a research paper

- Much longer and deeper

- Strong emphasis on theory and justification

- Requires original contribution to knowledge

- More focus on methodological rigor

- Discussion is more extensive and critical

Chapter 3 → How will I study it?

# What is Narrative Review Article?


A narrative review article is a type of academic writing that summarizes and interprets existing research on a topic in a descriptive and flexible way, without following the strict, protocol-driven methodology of a systematic review.

It is commonly used to:

- Explain a broad topic

- Trace the development of a field

- Critically discuss theories and concepts

- Provide context rather than exhaustive evidence synthesis

## What makes it different?

Unlike a systematic review, a narrative review:

- Does not follow a rigid search protocol

- Does not aim for complete coverage of all studies

- Relies more on the author’s expertise and interpretation

- Is more qualitative and conceptual

**That flexibility is useful, but it also means it can be biased if not done carefully.**

## Structure of a Narrative Review Article

There is no single fixed template, but a strong narrative review usually follows this logical structure:

### 1. Title

Clear and focused on the topic.

Example:
“Digital Transformation in Supply Chains: A Narrative Review”

### 2. Abstract

A concise summary including:

- Topic and scope

- Purpose of the review

- Key themes or insights

- Main conclusions

### 3. Introduction

Sets the stage.

- Includes:

- Background of the topic

- Importance of the issue

- Purpose of the review

- Scope (what is included and excluded)

**This is where we justify why a narrative review is needed.**

### 4. Body (Core of the Review)

This is the most important part and is organized thematically, not chronologically or methodologically.

Typical ways to structure it:

#### a. Thematic structure

- Theme 1 (e.g., digital technologies in supply chains)

- Theme 2 (e.g., risk and resilience)

- Theme 3 (e.g., performance outcomes)

#### b. Theoretical perspectives

- Different models or frameworks

- Competing viewpoints

#### c. Evolution over time (optional)

How the field has developed

Key requirement:
We must synthesize, not just summarize. That means:

- Compare studies

- Highlight agreements and contradictions

- Identify gaps

### 5. Discussion

Some papers merge this with the body, but when separate, it focuses on:

- Key insights across themes

- Research gaps

- Conceptual implications

- Future research directions

### 6. Conclusion

A concise wrap-up:

- Main takeaways

- Contribution of the review

- Final perspective on the topic

### 7. References

- All cited studies listed properly.

## Simple mental model

Introduction → What is the topic and why it matters?

Body → What does the literature say (organized by themes)?

Discussion → What do we learn from it?

Conclusion → What is the final takeaway?

# What is Systematic Review Article?

A systematic review article is a rigorous, method-driven study that collects, evaluates, and synthesizes all relevant research on a clearly defined question using a transparent and reproducible process.

Unlike a narrative review, this is not just “what I think the literature says.” It is:

- Protocol-based

- Comprehensive in search

- Explicit in inclusion/exclusion criteria

- Replicable by other researchers

In many fields, it is considered high-level evidence.

**Core idea:**

A systematic review answers a focused question like:

“What is the impact of supply chain disruptions on firm performance in emerging markets?”

Then it follows a structured pipeline:

- Define question

- Search all relevant studies

- Screen them

- Evaluate quality

- Synthesize findings

## Structure of a Systematic Review Article

Most systematic reviews follow standards like PRISMA (Preferred Reporting Items for Systematic Reviews and Meta-Analyses).

### 1. Title

Should clearly indicate it is a systematic review.

Example:
“A Systematic Review of Supply Chain Resilience and Firm Performance”

### 2. Abstract (Structured)

Usually divided into:

- Background

- Objectives

- Methods

- Results

- Conclusion

### 3. Introduction

Defines the foundation.

Includes:

- Background of the topic

- Research problem

- Clear research question (often using frameworks like PICO)

- Justification for the review

### 4. Methodology (Most Critical Section)

- This is what makes it “systematic.”

#### a. Review protocol

- Mention if registered (e.g., PROSPERO in some fields)

#### b. Search strategy

- Databases used (Scopus, Web of Science, etc.)

- Keywords and search strings

#### c. Inclusion and exclusion criteria

- Time period

- Type of studies

- Language

- Industry or context

#### d. Study selection process

- Screening steps (title → abstract → full text)

#### e. Data extraction

- What information was collected from each study

#### f. Quality assessment

- Tools or criteria used to evaluate study quality

**This section must be detailed enough that someone else can replicate our review.**

### 5. Results

Presents what was found from the selected studies.

#### a. Study selection summary

Often shown using a flow diagram (PRISMA flow):

- Number of studies identified

- Number screened

- Number included

#### b. Characteristics of studies

- Sample sizes

- Methods used

- Context (country, industry, etc.)

#### c. Findings synthesis

- Thematic synthesis OR

- Quantitative synthesis (meta-analysis, if applicable)

### 6. Discussion

- Interprets the findings.

- What patterns emerge?

- What is consistent or conflicting?

- Theoretical implications

- Practical implications

### 7. Limitations

- Limitations of included studies

- Limitations of your review process

### 8. Conclusion

- Key insights

- Contribution to knowledge

- Suggestions for future research

### 9. References
### 10. Appendices (optional)

- Search strings

- Coding tables

- Quality assessment tools

## Key difference vs Narrative Review

| Aspect        | Systematic Review          | Narrative Review |
| ------------- | -------------------------- | ---------------- |
| Approach      | Structured, protocol-based | Flexible         |
| Search        | Comprehensive              | Selective        |
| Bias          | Minimized                  | Higher risk      |
| Replicability | High                       | Low              |
| Purpose       | Answer specific question   | Provide overview |


## Simple mental model

Question → What exactly am I investigating?

Method → How did I systematically collect studies?

Results → What do the studies show?

Discussion → What does it mean overall?

## Important note (this is where many go wrong)

A systematic review is not just collecting papers and summarizing them.

If we don’t:

- Define clear criteria

- Document our search process

- Show screening steps

**then it is not a *systematic review*, even if we call it one.**

# What is a Research Paper?

A research paper is a formal academic document that presents original research, analysis, or findings on a specific question. Its purpose is to contribute new knowledge, test theories, or provide evidence-based insights.

In our domain, this usually means:

- Testing hypotheses (e.g., regression models in finance)

- Analyzing datasets (firm-level, macro, supply chain data)

- Drawing conclusions grounded in empirical or theoretical work

## Structure of a Research Paper

| Section                   | What it Includes                                                      | Key Purpose                                   |
| ------------------------- | --------------------------------------------------------------------- | --------------------------------------------- |
| **Title**                 | Clear, specific description of the study                              | Communicates the topic precisely              |
| **Abstract**              | Brief summary (purpose, method, results, conclusion)                  | Gives a quick overview of the entire paper    |
| **Keywords**              | 4–6 important terms                                                   | Helps indexing and discoverability            |
| **Introduction**          | Background, problem statement, research gap, objectives, hypotheses   | Sets context and justifies the study          |
| **Literature Review**     | Prior studies, theories, gaps                                         | Shows what is known and where your study fits |
| **Methodology (Methods)** | Research design, data sources, variables, models, tools               | Explains how the study was conducted          |
| **Results**               | Tables, figures, statistical outputs                                  | Presents findings without interpretation      |
| **Discussion**            | Interpretation of results, comparison with past studies, implications | Explains meaning and significance             |
| **Conclusion**            | Summary of findings, contributions, limitations, future research      | Wraps up the study                            |
| **References**            | All cited sources (APA, Harvard, etc.)                                | Ensures credibility and traceability          |
| **Appendices (optional)** | Survey tools, extra tables, technical details                         | Provides supporting material                  |


## Simple flow to remember

| Step              | Question it answers      |
| ----------------- | ------------------------ |
| Introduction      | What is the problem?     |
| Literature Review | What do we already know? |
| Methodology       | How did I study it?      |
| Results           | What did I find?         |
| Discussion        | What does it mean?       |
| Conclusion        | So what?                 |


## One practical insight

Most weak research papers fail not because of poor data, but because:

- The research gap is unclear, or

- The methodology is not aligned with the question, or

- The discussion does not connect back to theory

**If we get those three aligned, the paper becomes much stronger.**
